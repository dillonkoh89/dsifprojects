## Null value handling

```python
# Storing features that contain null values sorted by null value count
null_df = pd.DataFrame(data = housing_train[housing_train.columns[housing_train.isnull().any()]].isnull()
                       .sum().sort_values(ascending=False), columns = ['null_count'])
null_df['null_perc'] = null_df['null_count'].map(lambda x: round((x/len(housing_train))*100,2))
null_df
```

There are 26 fields with null values. On a high level, we will replace NaN with relevant values (after a preliminary profiling of the data where necessary):
1. For categorical feature, we will assume that the house do not have such features (e.g. fence, garage, basement) and replace NaN with values indicating none based on [data dictionary](http://jse.amstat.org/v19n3/decock/DataDocumentation.txt).
2. For numerical feature, we will replace NaN with mean / median / mode on case-by-case basis

```python
# Replacing NaN for categorical features
housing_train['pool_qc'] = housing_train['pool_qc'].fillna('NA') #replace NaN with NA (for no pool)
housing_train['misc_feature'] = housing_train['misc_feature'].fillna('NA') #replace NaN with NA (for none)
housing_train['alley'] = housing_train['alley'].fillna('NA') #replace NaN with NA (for no alley access)
housing_train['fence'] = housing_train['fence'].fillna('NA') #replace NaN with NA (for no fence)
housing_train['fireplace_qu'] = housing_train['fireplace_qu'].fillna('NA') #replace NaN with NA (for no fence)
housing_train['garage_cond'] = housing_train['garage_cond'].fillna('NA') #replace NaN with NA (for no garage)
housing_train['garage_qual'] = housing_train['garage_qual'].fillna('NA') #replace NaN with NA (for no garage)
housing_train['garage_finish'] = housing_train['garage_finish'].fillna('NA') #replace NaN with NA (for no garage)
housing_train['garage_type'] = housing_train['garage_type'].fillna('NA') #replace NaN with NA (for no garage)
housing_train['bsmt_exposure'] = housing_train['bsmt_exposure'].fillna('NA') #replace NaN with NA (for no basement)
housing_train['bsmtfin_type_2'] = housing_train['bsmtfin_type_2'].fillna('NA') #replace NaN with NA (for no basement)
housing_train['bsmt_cond'] = housing_train['bsmt_cond'].fillna('NA') #replace NaN with NA (for no basement)
housing_train['bsmt_qual'] = housing_train['bsmt_qual'].fillna('NA') #replace NaN with NA (for no basement)
housing_train['bsmtfin_type_1'] = housing_train['bsmtfin_type_1'].fillna('NA') #replace NaN with NA (for no basement)
housing_train['mas_vnr_type'] = housing_train['mas_vnr_type'].fillna('None') #replace NaN with None (for none)
```

```python
# Replacing NaN by inference
housing_train['mas_vnr_area'] = housing_train['mas_vnr_area'].fillna(0) #replace NaN with 0 (as the null count is matching with feature 'mas_vnr_type')
housing_train['bsmt_half_bath'] = housing_train['bsmt_half_bath'].fillna(housing_train['bsmt_half_bath'].mode()[0]) #replace NaN with mode (as not to have any bath in float value)
housing_train['bsmt_full_bath'] = housing_train['bsmt_full_bath'].fillna(housing_train['bsmt_full_bath'].mode()[0]) #replace NaN with mode (as not to have any bath in float value)
```

```python
# Replacing NaN 'garage_yr_blt' with 'year_built'
housing_train.loc[housing_train['garage_yr_blt'].isna(), ['garage_yr_blt']] = housing_train.loc[housing_train['garage_yr_blt'].isna()]['year_built']
```

```python
# Checking for record where basement area is null
housing_train[housing_train['total_bsmt_sf'].isnull()].groupby(by=["bsmt_qual",'total_bsmt_sf',
                          'bsmt_unf_sf','bsmtfin_sf_2','bsmtfin_sf_1'], dropna=False)['id'].count()
```

The null basement-related fields is of 1 record without any basement.

```python
# Replacing NaN numerical features related to basement
housing_train['total_bsmt_sf'] = housing_train['total_bsmt_sf'].fillna(0) #replace NaN with 0 (as the feature 'bsmt_qual' indicates no basement)
housing_train['bsmt_unf_sf'] = housing_train['bsmt_unf_sf'].fillna(0) #replace NaN with 0 (as the feature 'bsmt_qual' indicates no basement)
housing_train['bsmtfin_sf_2'] = housing_train['bsmtfin_sf_2'].fillna(0) #replace NaN with 0 (as the feature 'bsmt_qual' indicates no basement)
housing_train['bsmtfin_sf_1'] = housing_train['bsmtfin_sf_1'].fillna(0) #replace NaN with 0 (as the feature 'bsmt_qual' indicates no basement)
```

```python
# Checking for the remaining features with null values
housing_train[housing_train.columns[housing_train.isnull().any()]].isnull().sum().sort_values(ascending=False)
```

```python
# Distribution of 'lot_frontage'
plt.figure(figsize = (12,5))
plt.hist(housing_train['lot_frontage'], bins = 10)
plt.vlines(x = housing_train['lot_frontage'].mean(), ymin = 0, ymax = 1000, color = 'orange', label = 'Mean of lot frontage')
plt.title('Distribution of lot frontage')
plt.xlabel('Linear feet of street')
plt.ylabel('# of Houses')
plt.legend(loc = 'upper center');
```

## 2c. One-Hot Encoding for Categorical Data

```python
# Check datatypes
housing_train.info()
```

On a high level, the datatypes looks fine. 'id' and 'pid' should be excluded from modelling, as they are both identifiers which shouldn't be used as predictors.

Let's now apply the one-hot encoding to the categorical fields and generate the output into a new dataset.

```python
# Applying one-hot encoding to the categorical fields to consume as data points in machine learning model
to_encode = ['ms_subclass','ms_zoning','street','alley','lot_shape','land_contour','utilities','lot_config',
            'land_slope', 'neighborhood', 'condition_1', 'condition_2', 'bldg_type', 'house_style', 
             'roof_style','roof_matl', 'exterior_1st', 'exterior_2nd', 'mas_vnr_type',
             'exter_qual', 'exter_cond', 'foundation', 'bsmt_qual', 'bsmt_cond', 'bsmt_exposure', 
             'bsmtfin_type_1', 'bsmtfin_type_2', 'heating', 'heating_qc', 'central_air', 'electrical', 
             'kitchen_qual', 'functional', 'fireplace_qu', 'garage_type', 'garage_finish', 'garage_qual',
             'garage_cond', 'paved_drive', 'pool_qc', 'fence', 'misc_feature', 'sale_type']
housing_train_enc = housing_train #create a new dataset with encoded categorical variables

for i in to_encode:
    df = pd.get_dummies(housing_train_enc[i], prefix=i)
    housing_train_enc = pd.concat([housing_train_enc,df], axis=1)
    housing_train_enc.drop(i, axis=1, inplace=True)
housing_train_enc.head()
```

```python
housing_train_enc.shape
```

After encoding, we now have a dataset with 319 features.
